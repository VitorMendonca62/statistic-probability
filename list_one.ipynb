4. Em um estudo sobre os hábitos de consumo online, 2.500 usuários de uma plataforma
foram entrevistados. Queremos determinar a porcentagem desses usuários que preferem
compras online em relação às compras em lojas físicas. Qual seria o tamanho mínimo da
amostra aleatória simples necessário para garantir um erro amostral máximo de 2,5%?

```python
tolerable_sampling_error = 2.5 / 100
szie_population = 2500

first_approximation = 1 / (tolerable_sampling_error ** 2)

size_sample = (szie_population * first_approximation) / (szie_population + first_approximation)

print(f"The sample size is {size_sample:.0f}")
```

8) Os salários de 50 funcionários de uma empresa estão agrupados na tabela a seguir.
Determine a variância e o desvio padrão dos salários.

| Salário (mil reais) | Frequência |
|------|------ |
| 2−4 | 12 |
| 5−7 | 18 |
| 8−10 | 10 |
| 11−13 | 7 |
| 14−16 | 3 |

```python
table = {
  "2-4": 12,
  "5-7": 18,
  "8-10": 10,
  "11-13": 7,
  "14-16": 3
}

sum = 0
length = 0
averages_datas = []

for key in table:
  limit = key.split("-")
  limit_up = int(limit[0])
  limit_down = int(limit[1])
  
  Xi = (limit_up + limit_down) / 2
  averages_datas.append(Xi)
  
  length += table[key]
  sum += Xi * table[key]
  
average = sum / length

sum = 0
for i in range(len(averages_datas)):
  key = list(table.keys())[i]
  sum += ((averages_datas[i] - average) ** 2) * table[key]
  
print(f"The variance is {sum / length}")
print(f"The standard deviation is {(sum / length) ** (1/2)}")
  
  
```
